```python
 #Creamos una nueva columna 'date', asi manejamos la variable de tiempo para cada usuario posible.
events_training['date'] = pd.to_datetime(events_test['timestamp'])
```

```python
#este feature le agrega un peso a los usuarion segun la cantidad de conversiones que realizaron.
#Aquellos usuarion que han realizado una conversion tienen menos peso que aquellos que no la hicieron, ya que al haber
# comprado ya un producto es menos probable que ellos lo vuelvan a hacer.
#Ademas este peso se reduce si el usuario realizo mas de una conversion
datos_cuatro = events_training.loc[:,['person','event','timestamp']]
datos_cuatro = datos_cuatro.groupby(['person','event']).count().unstack().reset_index().fillna(0)
datos_cuatro.columns=['person','ad_campaign_hit','brand_listing','checkout','conversion','generic_listing',\
                      'lead','search_engine_hit','searched_products','staticpage','viewed_product','visited_site']
feature_uno = datos_cuatro
feature_uno['weight_conversion_per_person'] = datos_cuatro.loc[:,['person','conversion']]['conversion']\
            .apply(lambda x: 1/(x+1))
feature_uno = feature_uno.loc[:,['person','weight_conversion_per_person']]
feature_uno = label_train.merge(feature_uno,on='person',how='inner').drop('label',axis=1)
feature_uno.head()
```

```python
#vamos a crear un feature que es un peso que se le asigna a cada usuario de acuerdo a que tan proxima esta la
#ultima conversion que realizaron con respecto a la fecha limite donde empieza la primera semana de junio
#los usuarion que hayan realizado un conversion muy cerca a esta fecha tendran el peso menor.
# a medida que la ultima conversion del usuario se aleje de esta fecha el peso aumenta.
#si el usuario no realizo ninguna conversion entonces el peso es el maximo, que es 1 en este caso
fecha1 = events_training['date'].min()
fecha2 = pd.to_datetime('2018-05-31 23:59:59')
tiempo_total = (fecha2-fecha1).total_seconds()
datos_cinco = events_training.groupby('person').filter(lambda x: (x['event']=='conversion').any()).reset_index()\
            .loc[:,['person','event','date']]
datos_cinco_con = datos_cinco.loc[datos_cinco['event']=='conversion',:].groupby('person').max().reset_index()\
                .drop('event',axis=1)
datos_cinco_con['weight_conversion_per_time']= (fecha2 - datos_cinco_con['date']).dt.total_seconds()/tiempo_total
datos_cinco_con = datos_cinco_con.drop('date',axis=1)
datos_cinco_sin = events_test.groupby('person').filter(lambda x: (x['event']!='conversion').all()).reset_index()\
            .loc[:,['person','event']].drop_duplicates(subset='person',keep='first')
datos_cinco_sin['weight_conversion_per_time']=1
datos_cinco_sin = datos_cinco_sin.drop('event',axis=1)
feature_dos = datos_cinco_sin.merge(datos_cinco_con,on=['person','weight_conversion_per_time'],how='outer')
feature_dos = label_train.merge(feature_dos,on='person',how='inner').drop('label',axis=1)
feature_dos.head()
```
