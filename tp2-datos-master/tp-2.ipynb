```python
#calculamos cantidad de eventos que tuvo cada usuario
#maximo evento que tuvo
#minimo evento que tuvo
#media de eventos
```

```python
datos_uno = events_training.loc[:,['person','event']]
```

```python
cantidad_eventos = datos_uno.groupby('person').count().reset_index()
cantidad_eventos.columns=['person','event_count']
cantidad_eventos.head()
```

```python
datos_dos = events_training.loc[:,['person','event','timestamp']]
datos_dos = datos_dos.groupby(['person','event']).count().reset_index()
datos_dos.columns=['person','event','cant_per_event']
datos_dos.head()
```

```python
maximo_evento = datos_dos.groupby('person').apply(lambda x: x.nlargest(1,'cant_per_event')).reset_index(drop=True)
maximo_evento.head()
```

```python
minimo_evento = datos_dos.groupby('person').apply(lambda x: x.nsmallest(1,'cant_per_event')).reset_index(drop=True)
minimo_evento.head()
```

```python
minimo_evento.columns=['person','less_frequent_event','count_less_frequent_event']
```

```python
maximo_evento.columns=['person','most_frequent_event','count_most_frequent_event']
```

```python
media_evento = datos_dos.loc[:,['person','cant_per_event']]\
                .groupby('person').apply(lambda x: x.mean()).reset_index()
```

```python
media_evento = media_evento.drop('level_1',axis=1)
media_evento.columns=['person','mean_per_event']
```

```python
valores = media_evento.merge(maximo_evento,on='person',how='inner').merge(minimo_evento,on='person',how='inner')
```

```python
datos_tres = datos_dos.merge(cantidad_eventos,on='person',how='inner')
```

```python
datos_tres['mean_per_person'] = datos_tres['cant_per_event'].astype(float)/datos_tres['event_count']
```

```python
tasas_por_usuario = datos_tres.loc[:,['person','event','mean_per_person']]\
            .groupby(['person','event']).agg({'mean_per_person':lambda x: x}).unstack().reset_index().fillna(0)
```

```python
tasas_por_usuario.columns=['person','ad_campaign_hit_rate','brand_listing_rate','checkout_rate',\
                      'conversion_rate','generic_listing_rate','lead_rate','search_engine_hit_rate',\
                         'searched_products_rate','staticpage_rate','viewed_product_rate','visited_site_rate' ]
```
